```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn import datasets
import seaborn as sns
```

```python
X,y = datasets.make_regression(n_samples = 1000, n_features = 44,noise = 20, random_state = 443)
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state = 443, test_size = 0.2)
```

```python
plt.figure(figsize = (12,6))
plt.scatter(X[:,0],y,color = 'r',marker= 'o', s = 50)
plt.show()
```

```python
class LinearRegression():
    def __init__(self,lr = 0.001, n_iters = 10000):
        self.weights = None
        self.bias = None
        self.lr = lr
        self.n_iters = n_iters

    def fit(self,X,y):
        self.bias = 0
        n, n_features = X.shape
        self.weights = np.zeros(n_features)
        for _ in range(self.n_iters):
            y_pred = np.dot(X,self.weights) + self.bias
            dw = 1/n * np.dot(X.T,(y - y_pred))
            db = 1/n * np.sum(y - y_pred)
            self.weights -= self.lr * dw
            self.bias -= self.lr * db
    def predict(self,X):
        y_pred = np.dot(X,self.weights) + self.bias
        return y_pred
```

```python
lr = LinearRegression(lr = 0.01)
lr.fit(X_train,y_train)
```

```python
lr.weights,lr.bias
```

```python
y_pred = lr.predict(X_test)
y_pred
```
